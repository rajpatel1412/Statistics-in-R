```r
data.pesticide = read.csv(file = "pesticide.csv", header = TRUE)
print(data.pesticide)
```

```r
data.pesticide$proportion = data.pesticide$Pesticide/data.pesticide$Total
print(data.pesticide)
```

```r
barplot(height=data.pesticide$Pesticide, names.arg = data.pesticide$Food, 
       main = "Residual Pesticide in Grocery Store Foods", las = 2, ylim = c(0, 50),
       ylab = "Number of Items with Residual Pesticide")
```

```r
barplot(height = data.pesticide$proportion, names.arg = data.pesticide$Food,
        ylab = "Proportion of Foods with Residue",
       main = "Proporion of Foods with Residual Pesticide", las = 2, ylim = c(0, 1))
```
